```python
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
model = Word2Vec.load('./model/word2vec_battery.model')
```

```python
data = pd.read_csv('./data/元素周期表.txt', header=None, names=['元素'])
X = list(data['元素'])
words = model.wv.index2word
tmp = [val for val in X if val in words]
vectors = model[tmp]
```

```python
tsne = TSNE(n_components=2,early_exaggeration=12.0,learning_rate=200,n_iter=10000,init='pca')
embedd = tsne.fit_transform(vectors)
```

```python
def FenLei(s):
    i = 0
    x = []
    for a in tmp:
        for b in s:
            if a == b:
                x.append(embedd[i])
        i += 1
    y = np.array(x)
    return(y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(embedd[:,0], embedd[:,1], s=20)
for i in range(len(tmp)):
    x = embedd[i][0]
    y = embedd[i][1]
    plt.text(x, y, tmp[i], fontsize=10)
plt.show()
```
